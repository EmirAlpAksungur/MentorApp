```python
import torch
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Resmi yükle ve tespit yap
img_path = './data/images2/0a0bdec4b07ca3c1.jpg'  # Kullanmak istediğiniz resim dosyasının yolunu belirtin
results = model(img_path)

# Tespit edilen nesneleri ve sınırlayıcı kutuları göster
results.show()

# Sonuçları numpy dizisine dönüştür
detections = results.pandas().xyxy[0]  # (xmin, ymin, xmax, ymax, confidence, class, name)

# Tespit edilen sonuçları yazdır
print(detections)
```
